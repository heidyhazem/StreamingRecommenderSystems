```python
import pandas as pd
import glob
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

# ML 100K

```python
#centralized sgd (ISGD)
x_nodes = []
with open("recallSide.txt", 'r') as f:
    lines = f.readlines()
for l in lines:
    x_nodes.append(int(l.strip()))
recall_avgs_centralized = []
sma = []
for ct in x_nodes:
    sma.append(ct)
    n = len(sma)
    if n > 5000:
        del sma[0]
        n -= 1
        recall_avgs_centralized.append(sum(sma) / float(n))
```

```python
empty = pd.DataFrame(columns=["recall"])
path = "recall_*.txt"
for filename in glob.glob(path):
    nodes = filename.split("_")[1]
    x_nodes = []
    with open(filename, 'r') as f:
        lines = f.readlines()
    for l in lines:
        x_nodes.append(int(l.strip()))
    recall_avgs = []
    sma = []
    for ct in x_nodes:
        sma.append(ct)
        n = len(sma)
        if n > 5000:
            del sma[0]
            n -= 1
            recall_avgs.append(sum(sma) / float(n))
    empty[nodes] = recall_avgs
    
```

```python
empty
```

```python
empty = empty.drop(["recall"],axis = 1)
```

```python
results = list(empty.columns)
results = list(map(int, results))
sortedRes = sorted(results)
sortedRes = list(map(str, sortedRes))
empty = empty[sortedRes]
```

```python
fig, ax = plt.subplots()
#ax.plot(empty,ax=ax)
empty.plot(ax=ax)
ax.plot(recall_avgs_centralized,label = 'ISGD')
ax.set_xlabel('index')
ax.set_ylabel('recall@10')
ax.grid(True)
ax.set_title("SMA Recall for ISGD vs Parallel ISGD with different number of nodes using L=ML100K dataset")
#ax.set_xticks([0, 50000, 100000, 150000])
#ax.set_yticks([0.00, 0.05, 0.10, 0.15])
fig.set_size_inches((5,5))
fig.patch.set_alpha(0.5)
handles, labels = ax.get_legend_handles_labels()
ax.legend(labels)
```
