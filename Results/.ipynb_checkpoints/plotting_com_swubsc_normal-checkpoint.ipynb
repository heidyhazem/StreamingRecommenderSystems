```python
import pandas as pd
import glob
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#2 Equally distributed (PISGD)
x_nodes = []
with open("changing_ni_ml100/recall2.txt", 'r') as f:
    lines = f.readlines()
for l in lines:
    x_nodes.append(int(l.strip()))
recall_avgs_2 = []
sma = []
for ct in x_nodes:
    sma.append(ct)
    n = len(sma)
    if n > 5000:
        del sma[0]
        n -= 1
        recall_avgs_2.append(sum(sma) / float(n))
```
